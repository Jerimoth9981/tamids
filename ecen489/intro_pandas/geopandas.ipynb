```python
df = pd.read_csv("https://github.com/jtao/tamids/raw/master/ecen489/MJ_mass_shooting1982_2019.csv")
```

```python
df = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))
```

```python
df[['city','state']] = df.location.str.split(',', expand=True)
df['city'] = df['city'].str.strip()
df['state'] = df['state'].str.strip()
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
usa=usa = gpd.read_file('states_21basic/states.shp')
ax = usa[usa.STATE_NAME!='Alaska'].plot(figsize=(20,10),
    color='white', edgecolor='black')

# We can now plot our ``GeoDataFrame``.
df.plot(ax=ax, color='red')
plt.show()
```
